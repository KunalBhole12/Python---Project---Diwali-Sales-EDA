```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import csv file

df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
#drop unrelated/blank cloumns

df.drop(columns=['Status','unnamed1'],inplace=True)
df.head()
```

```python
#check for null values

df.isna().sum()
```

```python
df[df.Amount.isna()]
```

```python
na = df[df.Amount.isna()]
```

```python
#Grouping the DataFrame by 'Product_Category', and for each group, fill NaN values in the 'Amount' column with the mean of that particular group

df['Amount'] = df.groupby('Product_Category')['Amount'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.loc[na.index]
```

```python
df.dtypes
```

```python
#Replacing 1,0 values with Married,UnMarried values in the 'Marital_Status' column

df['Marital_Status'] = df['Marital_Status'].replace(1,'Married')
df['Marital_Status'] = df['Marital_Status'].replace(0,'UnMarried')
df.head()
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)

df.describe()
```

```python
# use describe() for specific columns

df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis


### Gender

```python
plt.style.use('default')
```

```python
# plotting a column chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('No. of Females and Males')
```

```python
# plotting a column chart for Total Sales by Gender

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue = 'Gender')
plt.title('Sales by Gender')
```

### Age

```python
# plotting a countplot chart to show distribution of Age Group

plt.figure(figsize=(10,5))
sns.countplot(x= df['Age Group'].sort_values())
plt.title('Distribution of Age Groups')
```

```python
# plotting a barplot to show sales by Age Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')
plt.title('Sales by Age Group')
```

### State

```python
# plotting a barplot to show Top 10 States by sales

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True).tail(10)

plt.figure(figsize=(10,5))
plt.barh('State' , width='Amount' , data = sales_state, color = 'green')
plt.title('Top 10 States by Sales')
plt.xlabel('Sales')
plt.ylabel('State')
```

### Martial Status

```python
#plotting a column chart to show No. of Married and UnMarried Customers

ax = sns.countplot(data = df, x = 'Marital_Status' , hue='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('No. of Married and UnMarried Customers')
```

```python
#plotting a barplot to show sales by marital status

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.title('Sales by Marital Status')
```

### Occupation

```python
#plotting a countplot to show distribution of customers by occupation

plt.figure(figsize=(20,5))
ax = sns.countplot(data = df, x = 'Occupation', hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Distribution of Customers by Occupation')
```

### Product Category

```python
#plotting a bar graph to show Top 10 Categories by Sales

sales_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True).tail(10)

plt.figure(figsize=(10,5))
plt.barh('Product_Category' , width='Amount' , data = sales_category, color = 'red')
plt.title('Top 10 Categories by Sales')
plt.xlabel('Category')
plt.ylabel('State')
```

### Products

```python
#plotting a bar graph to show Top 10 Products Sold

product_orders = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x='Orders' , y='Product_ID' , data = product_orders, hue='Product_ID')
plt.title('Top 10 Products Sold')
plt.xlabel('Orders')
plt.ylabel('Product_ID')
```
